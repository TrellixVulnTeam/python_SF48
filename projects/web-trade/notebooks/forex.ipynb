```python
import pandas as pd
import numpy as np
import sqlite3
from datetime import datetime
import plotly
import plotly.graph_objects as go
import plotly.express as px

class Bot(object):
    def __init__(self, ticker, start, end, time_frame):
        self.ticker = ticker
        self.start = start
        self.end = end
        self.time_frame = time_frame
        
    def load_data(self, ticker, time_frame): # Load dataset.
        params_dic = {
         'host': 'localhost',
         'database': 'data/forex_historical.db'
        }
        connection = sqlite3.connect(f"{params_dic['database']}")
        dataset = pd.read_sql_query("SELECT * FROM "+ticker.lower()+"_"+time_frame, connection)
        dataset = dataset.astype({'time':'datetime64'})
        dataset.index = dataset['time']
        connection.close()
        return pd.DataFrame(dataset)


if __name__ == '__main__':
    #time_frames = {'mt5.TIMEFRAME_H12':'h12', 'mt5.TIMEFRAME_D1':'d1', 'mt5.TIMEFRAME_W1':'w1', 'mt5.TIMEFRAME_MN1':'mn1'}
    #tickers = ['AUDCAD', 'AUDNZD', 'AUDUSD', 'CADCHF', 'EURAUD', 'EURCAD', 'EURCHF', 'EURHUF', 'EURGBP', 'EURJPY', 'EURNZD', 'EURUSD', 'GBPCAD', 'GBPCHF', 'GBPJPY', 'GBPUSD', 'GBPNZD', 'USDCAD', 'USDCHF', 'USDJPY', 'USDMXN', 'USDSGD', 'USDTRY', 'NZDCAD', 'NZDUSD', 'NZDJPY', 'XAGUSD', 'XAUUSD']
    time_frames = {'mt5.TIMEFRAME_H12':'h12'}
    tickers = ['AUDCAD', 'AUDNZD']
    for ticker in tickers:
        print('\n')
        for time_frame in time_frames:
            forex = Bot(ticker, datetime(2021, 12, 30), datetime(2021, 12, 31), time_frames[time_frame]).load_data(ticker, time_frames[time_frame])
            print(forex)
            fig = go.Figure(
                data=[go.Candlestick(x=forex.index, low=forex['low'], high=forex['high'], close=forex['close'], open=forex['open'], increasing_line_color='green', decreasing_line_color='red')]
            )
            fig.show()
```
